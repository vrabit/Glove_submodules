```python
#load model and scaler, change model name if using other pickeled models
import pickle
from sklearn.preprocessing import StandardScaler
from sklearn.multiclass import OneVsRestClassifier
from sklearn.neighbors import KNeighborsClassifier

#predict
from serial import*
from collections import deque 

#send data
import asyncio

#load models
file_path = "best_knn_model.pkl"
with open(file_path, 'rb') as f:
    loaded_model = pickle.load(f)
    
file_path = "standard_scaler.pkl"
with open(file_path, 'rb') as f:
    scaler = pickle.load(f)

#open glove port
serialPort = 'COM4'
baudRate = 19200
ser = Serial(serialPort,baudRate,timeout=1)


async def connect_to_server(loaded_model, scaler, ser):
    reader, writer =  await asyncio.open_connection('127.0.0.1',8877)
    print('connection with server established')
    #data buffer
    buf = deque()

    data = ser.readline()
    first = True
    while True:
        data = ser.readline()
        #GyroXRaw = GyroYRaw = GyroZRaw = 0
        flexThumb = flexIndex = flexMiddle = flexRing = flexPinky = 0

        try:
            GyroXRaw = data[2:8].decode('utf-8')
            GyroYRaw = data[10:16].decode('utf-8')
            GyroZRaw = data[18:23].decode('utf-8')
            flexThumb = int(data[27:33].decode('utf-8'))
            flexIndex = int(data[37:43].decode('utf-8'))
            flexMiddle = int(data[47:53].decode('utf-8'))
            flexRing = int(data[57:63].decode('utf-8'))
            flexPinky = int(data[67:73].decode('utf-8'))
                
            #push list into queue to be processed
            flexList = [flexThumb, flexIndex, flexMiddle, flexRing, flexPinky]
            buf.append(flexList)
    
            if first == True:
                first = False
                Element_Input_standardized = scaler.fit_transform(buf)
            else:
                Element_Input_standardized = scaler.transform(buf)
    
            
            Element_pred_test = loaded_model.predict(Element_Input_standardized)
            #print(Element_pred_test[0])
            buf.pop()
    
            message = str(Element_pred_test[0]) + " " + GyroXRaw + " " + GyroYRaw + " " + GyroZRaw
            writer.write(message.encode())
            await writer.drain()
            data = await reader.read(1024) # Receive back from server (optional)
            print(f"Received from server: {data.decode()}")
    
        except Exception as e:
            print(f"Excetion: {type(e)}")  



try:
    await connect_to_server(loaded_model, scaler, ser)
except KeyboardInterrupt:
    print("keyboard interrupt")
finally:
    ser.close()
```
